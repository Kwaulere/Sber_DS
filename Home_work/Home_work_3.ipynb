```python
class Task:
    @staticmethod
    def avg_region_price(df):
        
        avg_region_price = df.groupby('data.developer.regRegionDesc').mean().sort_values(by='data.objPriceAvg')
        avg_region_price = avg_region_price['data.objPriceAvg']
        avg_region_price.dropna(axis=0, inplace=True)
        rcParams['figure.figsize'] = 9,20
        avg_region_price.plot(kind='barh')
        plt.xlabel("Средняя цена за кв метр",fontsize=14, fontweight="bold")
        plt.ylabel("Регион",fontsize=14, fontweight="bold")
        avg_region_price.sort_values(ascending=True)
        plt.show()
   
    @staticmethod
    def volume_of_apartments(df):
        
        df["data.objReady100PercDt"] = pd.to_datetime(df["data.objReady100PercDt"], format="%Y-%m-%d").dt.year
        df_perc = df[["data.developer.regRegionDesc", "data.objReady100PercDt"]]
        
    
        return df_perc.groupby(['data.objReady100PercDt']).size()
    
    @staticmethod
    def floor(df):
        
        floor_mean = df[["data.developer.regRegionDesc", 'data.floorMax']].groupby(["data.developer.regRegionDesc"],as_index=False).mean().sort_values(by="data.floorMax")
        sns.set(font_scale=3)
        plt.figure(figsize=(40,20))
        bar = sns.barplot(x="data.developer.regRegionDesc", y='data.floorMax', data=floor_mean)
        bar.set_xticklabels(bar.get_xticklabels(),rotation = 90)
        bar.set_xlabel("Регион", fontsize=40)
        bar.set_ylabel('Средняя этажность', fontsize=40)
        bar.set_title('Сравнение этажности домов по регионам')
        
    @staticmethod
    def count(df):
        count_flat = df[['data.developer.regRegionDesc', 'data.objElemLivingCnt']].groupby('data.developer.regRegionDesc', as_index=False).sum().sort_values('data.objElemLivingCnt', ascending=True)
        
        sns.set(font_scale=3)
        plt.figure(figsize=(40,20))
        bar = sns.barplot(x="data.developer.regRegionDesc", y='data.objElemLivingCnt', data=count_flat)
        bar.set_xticklabels(bar.get_xticklabels(),rotation = 90)
        bar.set_xlabel("Регион", fontsize=40)
        bar.set_ylabel('Количество квартир', fontsize=40)
        bar.set_title('Сравнение количества квартир')

    @staticmethod
    def square(df):
        
        df['MeenSquare'] = df['data.objSquareLiving']/df['data.objElemLivingCnt']
        flat_square = df[['data.developer.regRegionDesc', 'MeenSquare']].dropna()
        flat_square = flat_square.groupby('data.developer.regRegionDesc',as_index=False).mean().sort_values(by='MeenSquare', ascending=False)

        sns.set(font_scale=3)
        plt.figure(figsize=(40,20))
        bar = sns.barplot(x="data.developer.regRegionDesc", y='MeenSquare', data=flat_square)
        bar.set_xticklabels(bar.get_xticklabels(),rotation = 90)
        bar.set_xlabel("Регион", fontsize=40)
        bar.set_ylabel('Площадь', fontsize=40)
        bar.set_title('Средняя площадь квартир по регионам')

        
```
